```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### 折线图（plot）

```python
import numpy as np
x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
# plot绘制折线图
plt.plot(x, siny)
plt.plot(x, cosy)
# show展示图形
plt.show()
```

```python
# color修改颜色
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

```python
# linestyle修改样式
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="-.")
plt.show()
```

```python
# xlim修改x轴范围，ylim修改y轴范围
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
# axis修改各个坐标范围
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
# xlabel设置x轴label，ylabel设置y轴label
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# label设置线条label（需要调用legend）
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
# title设置图形标题
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("title")
plt.show()
```

### 散点图（Scatter Plot）

```python
plt.scatter(x, siny)
plt.scatter(x, cosy)
plt.show()
```

```python
# 散点图的x,y轴一般都用于表示（二维）特征，点颜色表示样本分布
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y)
plt.show()
```

```python
# alpha表示点透明度，color表示点颜色，marker表示点样式
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.1, color="red", marker="x")
plt.show()
```

### sklearn机器学习库

```python
# 加载数据集模块
from sklearn import datasets
```

```python
# 鸢尾花数据集
iris = datasets.load_iris()
iris.keys()
```

```python
print(iris.DESCR)
```

```python
X = iris.data
y = iris.target
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1], color="red", marker="o")
plt.scatter(X[y==1, 0], X[y==1, 1], color="blue", marker="+")
plt.scatter(X[y==2, 0], X[y==2, 1], color="green", marker="x")
plt.show()
```
